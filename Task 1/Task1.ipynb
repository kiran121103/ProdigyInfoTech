```python
# Drop unuseful columns
df.drop(columns=['Country Code','Indicator Name'],inplace=True)
```

```python
# filter data for total population
total_population_df=df[df['Indicator Code']=='SP.POP.TOTL']
total_population_df
```

```python
# sort data based on total population for 2023
total_population_sort = total_population_df.sort_values(by='2023',ascending=False)
total_population_sort
```

## get top 10 countries with the highest population for 2023

```python
# get top 10 countries with the highest population for 2023
top_ten_countries = total_population_sort.head(10)
print("Top Ten Counteies Of Total Population:-")
print(top_ten_countries)
```

## Data visualization

```python
# Data visualization
def add_value_labels(ax,spacing=10):
    for rect in ax.patches:
        x=rect.get_x()+rect.get_width()/2
        y=rect.get_height()-3
        
        va='bottom' if y>=0 else 'top'
        label= '()'.format(y) 
        y_shift = spacing * (1 if y >= 0 else -1)
        ax.annotate(label,(x,y),xytext=(0,y_shift),textcoords="offset points",ha='center',va=va)
```

## Top ten counteirs of total population in year 2023 and 2000

```python
# creat a bar plot
plt.figure(figsize=(20,13))
plt.rcParams['axes.facecolor']=="black"

plt.subplot(2,2,1)
plot=sns.barplot(x='2023',y="Country Name",data=top_ten_countries,palette="coolwarm")
plt.title("Top Ten countries Of Total Population (2023)")
plt.xlabel("Total population")
plt.ylabel("country")
add_value_labels(plot)

plt.subplot(2,2,2)
plot=sns.barplot(x='2000',y="Country Name",data=top_ten_countries,palette="coolwarm")
plt.title("Top Ten countries Of Total Population in 2000")
plt.xlabel("Total population")
plt.ylabel("country")
add_value_labels(plot)

plt.tight_layout()
```
